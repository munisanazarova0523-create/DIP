```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/Users/230070/Downloads/bird.png", 0) 
```

```python
def gamma_correction(I_in, c, gamma):
    I_norm = I_in.astype('float32') / 255.0
    I_gamma = c * (I_norm ** gamma)
    I_out = np.uint8(I_gamma * 255)

    return I_out
```

```python
gamma = 0.5
c = 1

result = gamma_correction(img, c, gamma)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("original")
plt.imshow (img, cmap= 'gray')

plt.subplot(1,2,2)
plt.title("gamma corrected")
plt.imshow(result,cmap='gray')

plt.show()
```

```python
cv2.imwrite("/Users/230070/Downloads/bird.png", result)
```

```python
#part2
def contrast_stretching(I_in, smin, smax) :
    rmin = np.min(I_in)
    rmax = np.max(I_in)

    stretched = ((smax - smin)/(rmax - rmin)) * (I_in - rmin) + smin

    return np.uint8(stretched)
```

```python
cs = contrast_stretching(img, 0, 255)

plt.imshow(cs, cmap='gray')
plt.title("Contrast Stretched")
plt.axis('off')
```

```python
img_norm = img.astype('float32') / 255

cs_norm = contrast_stretching(img_norm, 0, 1)

plt.imshow(cs_norm, cmap='gray')
plt.axis('off')
```

- Gamma < 1 brightens dark images and reveals hidden details.

Contrast Stretching
- Expands intensity range and improves contrast.

Normalization Effect
- Normalization stabilizes calculations but visual output remains similar after rescaling.
