```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(cv2.__version__)
```

```python
#task 1
image = cv2.imread('/Users/230070/Downloads/bird.png')
(h, w) = image.shape[:2]

# Example 1
T1 = np.float32([[1, 0, 100],
                 [0, 1, 50]])
translated1 = cv2.warpAffine(image, T1, (w, h))

# Example 2
T2 = np.float32([[1, 0, -80],
                 [0, 1, -40]])
translated2 = cv2.warpAffine(image, T2, (w, h))

plt.imshow(cv2.cvtColor(translated1, cv2.COLOR_BGR2RGB))
plt.title("Translated (Right 100, Down 50)")
plt.axis("off")
plt.show()


plt.imshow(cv2.cvtColor(translated2, cv2.COLOR_BGR2RGB))
plt.title("Translated (Right -80, Down -40")
plt.axis("off")
plt.show()
```

```python
#task 2
(h, w) = image.shape[:2]
center = (w // 2, h // 2)

R1 = cv2.getRotationMatrix2D(center, 90, 1.0)
rotated90 = cv2.warpAffine(image, R1, (w, h))

R2 = cv2.getRotationMatrix2D(center, 180, 1.0)
rotated180 = cv2.warpAffine(image, R2, (w, h))

R3 = cv2.getRotationMatrix2D(center, 45, 0.5)
rotated_scaled = cv2.warpAffine(image, R3, (w, h))

plt.imshow(cv2.cvtColor(rotated90,
cv2.COLOR_BGR2RGB))
plt.title('Rotated Image (90 degrees)')
plt.axis('off')
plt.show()
print(rotated90.shape)

plt.imshow(cv2.cvtColor(rotated180,
cv2.COLOR_BGR2RGB))
plt.title('Rotated Image (180 degrees)')
plt.axis('off')
plt.show()
print(rotated180.shape)

plt.imshow(cv2.cvtColor(rotated_scaled,
cv2.COLOR_BGR2RGB))
plt.title('Rotated Image (45 degrees with scaling 0.5)')
plt.axis('off')
plt.show()
print(rotated_scaled.shape)



```

```python
#task 3
Factor = 2
scaling_matrix = cv2.getRotationMatrix2D((0, 0), 0,
Factor)

scaled_image = cv2.warpAffine(image,
scaling_matrix, (w*Factor, h*Factor))

plt.imshow(cv2.cvtColor(scaled_image,
cv2.COLOR_BGR2RGB))
plt.title('Scaled Image (2x)')
plt.axis('off')
plt.show()
print(scaled_image.shape)
```

```python
#task 4
m = 0.67
shearing_matrix = np.float32([[1, m, 0], [0, 1,
0]])

sheared_image = cv2.warpAffine(image,
shearing_matrix, (w, h))

plt.imshow(cv2.cvtColor(sheared_image,
cv2.COLOR_BGR2RGB))
plt.title('Sheared Image (x-axis, factor ' + str(m)
+')')
plt.axis('off')
plt.show()
```

```python
m = 0.5
shear_x = np.float32([[1, m, 0],
                      [0, 1, 0]])

sheared_x = cv2.warpAffine(image, shear_x, (w, h))

m = 0.5
shear_y = np.float32([[1, 0, 0],
                      [m, 1, 0]])

sheared_y = cv2.warpAffine(image, shear_y, (w, h))

plt.imshow(cv2.cvtColor(sheared_x,
cv2.COLOR_BGR2RGB))
plt.title('Sheared Image (x-axis, factor ' + str(m)
+')')
plt.axis('off')
plt.show()

plt.imshow(cv2.cvtColor(sheared_y,
cv2.COLOR_BGR2RGB))
plt.title('Sheared Image (x-axis, factor ' + str(m)
+')')
plt.axis('off')
plt.show()


```
